```python
import sys

# Creating a file and handling exception in case the file already exists
try:
    f = open("petdata.dat","x").close()
    
except FileExistsError:
    print()

# Pet Class is created to get and set pet details
# It is used as the parent class for Fish, Mammal and Ambhibian 
class Pet:
    def __init__(self,petName,Age,bweight,name,address):
        self.__petName=petName
        self.__age=Age
        self.__bweight=bweight
        self.__owner=Person(name, address)
    def get_petName(self):
        return self.__petName
    def set_petName(self,petName):
        self.__petName=petName
        
    def get_age(self):
        return self.__age
    def set_age(self,Age):
        self.__age=Age
        
    def get_bweight(self):
        return self.__bweight
    def set_bweight(self,bweight):
        self.__bweight=bweight
        
    def get_owner(self):
        return self.__owner
    def set_owner(self,owner):
        self.__owner=owner
        
    def __repr__(self):
        s=self.get_petName() +';'+str(self.get_age())+';'+ str(self.get_bweight())+';'+ repr(self.get_owner())
        return s
    
    def __str__(self):
        s='Pet Name:' + self.get_petName() +'; '
        s+='Pet Age:' + str(self.get_age()) + ' days old; '
        s+='Birth Weight:' + str(self.get_bweight())
        s+="\n"+str(self.get_owner())
        
        return s
    
# Person Class is created to get and set person details

class Person:
    def __init__(self,name, address):
        self.__name=name
        self.__address=address
            
    def get_name(self):
        return self.__name
    def set_name(self,name):
        self.__name=name   
            
    def get_address(self):
        return self.__address
    def set_address(self,address):
        self.__address=address 
        
    def __repr__(self):
        s=self.get_name()+";"+self.get_address()
        return s
    
    def __str__(self):
        s='Owner Name: ' +self.get_name()+ '; '
        s+='Owner Address: ' +self.get_address()
        return s

# Mammal class which is a child class of Pet Class to get and set Mammal details 
# Weight function is defined to calculate the weight based on birth weight and age using the logic -
# Weight increases by 10% every 60 days for the first 300 days. Weight is constant after that. 
# Assume that the increase occurs on the 60th day only

class Mammal(Pet):
    def __init__(self,petName,Age,bweight,name,address,lsize,claw):
        super().__init__(petName,Age,bweight,name,address)
        self.__lsize=lsize
        self.__claw=claw
        #self.__age=Age
        #self.__weight=bweight
        
    def get_lsize(self):
        return self.__lsize
    def set_owner(self,lsize):
        self.__lsize=lsize   
            
    def get_claw(self):
        return "Yes" if(self.__claw) else "No"
    def set_claw(self,claw):
        self.__claw=claw
    def weight(self):
        a= self.get_age()
        b= self.get_bweight()
        if a >= 300:
            x = 5
        else:
            x=a//60

        for i in range(1,x+1):
            b +=(b*.10)

        return b
            
    def __repr__(self):
        s= ';'+str(self.get_lsize())+';'+self.get_claw()
        return '1;'+super(Mammal, self).__repr__() +s
    
    def __str__(self):
        s="\nLitter Size: "+str(self.get_lsize())+"; Has Claws? "+ self.get_claw()
        return super(Mammal, self).__str__() +s

# Fish class which is a child class of Pet Class to get and set Fish details 
# Weight function is defined to calculate the weight based on birth weight and age
# Weight increases by 5% every 80 days for the first 240 days and then stays constant. 
# Assume that the increase occurs on the 80th day only

class Fish(Pet):
    def __init__(self,petName,Age,bweight,name,address,scondition,length):
        super().__init__(petName,Age,bweight,name,address)
        self.__scondition=scondition
        self.__length=length
        
    def get_scondition(self):
        return self.__scondition
    def set_scondition(self,scondition):
        self.__scondition=scondition   
            
    def get_length(self):
        return self.__length
    def set_length(self,length):
        self.__length=length
    def weight(self):
        
        if self.get_age() >= 240:
            x = 3
        else:
            x=self.get_age()//80
        
        self.__weight = self.get_bweight()
        for i in range(1,x+1):
            self.__weight =self.__weight+(self.__weight*.05)

        return (self.__weight)
        
    def __repr__(self):
        s= ';'+self.get_scondition()+';'+str(self.get_length())
        return '2;'+super(Fish, self).__repr__() +s
    
    def __str__(self):
        s="\nScale condition: "+self.get_scondition().capitalize() +"; Length: "+str(self.get_length())
        return super(Fish, self).__str__() +s

# Amphibian class which is a child class of Pet Class to get and set Amphibian details 
# Weight function is defined to calculate the weight based on birth weight and age
# Weight increases by 5% every 120 days for the first 360 days and then increases by 3% every 120 days. 
# Assume that the increase occurs on the 120th day only
   
class Amphibian(Pet):
    
    def __init__(self,petName,Age,bweight,name,address,Venomous):
        super().__init__(petName,Age,bweight,name,address)
        self.__Venomous=Venomous
                
        
    def get___Venomous(self):
        return self.__Venomous
    def set___Venomous(self,Venomous):
        self.__Venomous=Venomous
        
    def weight(self):
        self.__weight = self.get_bweight()
        if self.get_age() <= 360:
            x = self.get_age()//120
            for i in range(1,x+1):
                self.__weight =self.__weight+(self.__weight*.05)
            return (self.__weight)
        else:
            a=self.get_age()-360
            for i in range(1,4):
                self.__weight =self.__weight+(self.__weight*.05) 

            y= a//120
            for i in range(1,y+1):
                self.__weight =self.__weight+(self.__weight*.03)
            return (self.__weight)
    
    def __repr__(self):
        s= ';'+str(self.get___Venomous())
        return '3;'+super(Amphibian, self).__repr__() +s
    
    def __str__(self):
        s="\nIs Venomous: "+str(self.get___Venomous())
        return super(Amphibian, self).__str__() +s
    
# add_pet function is defined to add the pet details entered by users into the file        
def add_pet(s):
    f = open("petdata.dat","a")
    f.write(s)
    f.write('\n')
    f.close()

# display_pets function is defined to display the pet details and pet weight from the file based on the option selected by the user 
# In case the option is 2 weights of the pets are displayed
# In case the option is 3 all the details in the file are displayed
def display_pets(option):
    f = open("petdata.dat","r")
    first = f.read(1)
    if not first:        
        print("No pets in inventory!")
    else:
        f.seek(0)
        for i in f:
            if(len(i.split(";")) > 2):
                if(int(i.split(';')[0]) == 1):
                    mam=Mammal(i.split(';')[1], int(i.split(';')[2]), float(i.split(';')[3]), i.split(';')[4], i.split(';')[5], int(i.split(';')[6]), i.split(';')[7])
                    if(option == 2):
                        print(mam.get_petName()+" weighs "+ str(round(mam.weight()))+" pounds")
                    else:
                        print("\nDetails of this Mammal pet are:\n"+str(mam))
                elif(int(i.split(';')[0]) == 2):
                    fish=Fish(i.split(';')[1], int(i.split(';')[2]), float(i.split(';')[3]), i.split(';')[4], i.split(';')[5], i.split(';')[6], int(i.split(';')[7]))
                    if(option == 2):
                        print(fish.get_petName()+" weighs "+ str(round(fish.weight()))+" pounds")
                    else:
                        print("\nDetails of this Fish pet are:\n"+str(fish))
                elif(int(i.split(';')[0]) == 3):
                    amp=Amphibian(i.split(';')[1], int(i.split(';')[2]), float(i.split(';')[3]), i.split(';')[4], i.split(';')[5], i.split(';')[6])
                    if(option == 2):
                        print(amp.get_petName()+" weighs "+ str(round(amp.weight()))+" pounds")
                    else:
                        print("\nDetails of this Amphibian pet are:\n"+str(amp))
        
# Function IsVenomous to validate and ensure that the user input is either 'Yes' or 'NO'
def ret_bool(string):
    s=str(input(string))
    if (s.lower() in ['true', '1', 't', 'y', 'yes', 'yeah', 'yup', 'certainly', 'uh-huh']):
        return True
    elif (s.lower() in ['false', '0', 'f', 'n', 'no', 'nope', 'nah', 'uncertainly']):
        return False
    else:
        print("invalid input")
        ret_bool(string)
    
    
# input_details function is defined to get the input from the users
# Here an assumption has been made that age and weight will not be 0 or less. In case User Enter 0 or lesser values error
# message will be thrown and user will be prompted to enter correct details
# In case User Enter a non numeric value for Age(string or float) or Weight(string), the program will throw error message and exit the program as 
# per the requirement
# Validations have been added to ensure user enters only Yes or No answer for hasclaw. Else error message is displayed 
# and user is prompted to re enter the right input
        
def input_details(j):
    try:
        petName=str(input("Enter the pet's name: "))
        try:
            Age=int(input("Enter the pet's age: "))
            if (Age<=0):
                raise NameError
            bweight=float(input("Enter the pet's birth weight: "))
            if (bweight<=0):
                raise NameError
        except ValueError:
            print('Only non numeric positive number should be entered. Exiting the application')
            sys.exit(0)
                                 
        except NameError:
            print("Invalid Input. Enter numeric value greater than 0")
            input_details(j)
        else:
            name=str(input("Enter the owner's name: "))
            address=str(input("Enter the owner's address: "))
            if j==1:
                lsize=int(input("Enter the litter size: "))
                claw=ret_bool("Does the pet have claws? Answer 'yes' or 'no': ")
                mam=Mammal(petName,Age,bweight,name,address,lsize,claw)
                print(mam)
                add_pet(repr(mam))
            
            # Validations has been added to ensure user enters either scaly or rough for scale condition and nothing else
            # In case of wrong input user will be prompted to re enter the input
            elif j==2:
                length=int(input("Enter the length of the fish in inches: "))
                def sconditioninput():
                    print("What is the condition of the scale? : ") 
                    try:
                        def sconditionval (scondition):      
                            if scondition.lower()=='rough' or scondition.lower()=='scaly':
                                fish=Fish(petName,Age,bweight,name,address,scondition,length)
                                print(fish)
                                add_pet(repr(fish))
                            else: 
                                raise ValueError
                            return scondition
                        scondition=str(input("Answer 'rough' or 'scaly': "))
                        sconditionval(scondition)
                    
                    except ValueError:
                        print ('Incorrect Input')
                        sconditioninput()
                    except Exception as e:
                        print ('Incorrect Input',e)
                        sconditioninput()
                    
                sconditioninput()
                
            # Validations have been added to ensure user enters only Yes or No answer for venomous. Else error message is displayed 
            # and user is prompted to re enter the right input 
            
            elif j==3:
                Venomous=ret_bool("Is the pet venomous? Answer 'yes' or 'no':")
                ambp=Amphibian(petName,Age,bweight,name,address,Venomous)
                print(ambp)
                add_pet(repr(ambp))
            
    except ValueError:
        print("Invalid Input")
        input_details(j)
        
    except Exception as e:
        print("Invalid Input", e)
        input_details(j)

# pet_input function is defined to get the type of pet details that they want to enter from the users
def pet_input(pet_type):
    if pet_type not in (1,2,3):
            print("Invalid Input. Kindly select the option between 1- 3 from the below menu")
            print()
            print("Please enter the type of Pet from the below option \n 1. Mammal \n 2. Fish \n 3. Amphibian")
            pet_type=int(input("Please enter your selection: "))
            pet_input(pet_type)
    else:
        return pet_type

    
def main():
    print("=====MENU=====")
    print("1. To add a pet ")
    print("2. To print current weight of all pets")
    print("3. To print all pets and owners")
    print("4. To exit program ")
    
    try:
        n =int(input("Your selection: "))
    except ValueError:
        print()
        print("Invalid Input.Please enter an number value between 1-4 from the below menu")
        print()
        main()
    except Exception as e:
        print()
        print("Invalid Input.",e)
        print()
        main()
    else:
        if n not in (1,2,3,4):
            print("Invalid Input. Kindly select the option between 1- 4 from the below menu")
            print() 
            main()
        else:

            if n ==1:
                try:
                    pet_type=int(input("Type of pet? (1. Mammal; 2. Fish; 3. Amphibian) : "))
                    pet_type=pet_input(pet_type)
                    input_details(pet_type)
                except ValueError:
                    print()
                    print("Invalid Input.Please enter an number value between 1-3 from the below option")
                    print()
                    print("Please enter the type of Pet from the below option \n 1. Mammal \n 2. Fish \n 3. Amphibian")
                    pet_type=int(input("Please enter your selection: "))
                    pet_type=pet_input(pet_type)
                    input_details(pet_type)
                #else:
                #    if pet_type not in (1,2,3):
                #        print("Invalid Input. Kindly select the option between 1- 3 from the below menu")
                #        print() 
                #        pet_type=int(input("Please enter the type of Pet from the below option \n 1. Mammal \n 2. Fish \n 3. Amphibian"))
                #    else:
                except Exception as e:
                    print()
                    print("Invalid Input",e)
                    print("Please enter an number value between 1-3 from the below option")
                    print()
                    print("Please enter the type of Pet from the below option \n 1. Mammal \n 2. Fish \n 3. Amphibian")
                    pet_type=int(input("Please enter your selection: "))
                    pet_type=pet_input(pet_type)
                    input_details(pet_type)
                        
                    
                main()



            elif n == 2:
                print("PetName and Current Weight of All pets")
                print("========================================")
                display_pets(n)
                main()


            elif n == 3:
                display_pets(n)
                main()



            else:
                def exitinput():
                    print("You chose to exit the program")
                    print("Are you sure you want to Exit the program ? (Y/N) : ") 
                    try:
                        def exitapp(i):      
                            if i.lower()=='y':
                                sys.exit(0)
                            elif i.lower()=='n':
                                main()
                            else: 
                                raise ValueError
                        i = str(input("Please enter(Y/N) : "))
                        exitapp(i)
                    except SystemExit:
                        print('Existing the application ')
                    except ValueError:
                        print ('Incorrect Input')
                        exitinput()
                    except Exception as e:
                        print ('Incorrect Input',e)
                        exitinput()
                exitinput()
            
                    

print()
main()
```
